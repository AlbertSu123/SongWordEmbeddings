# Example Response from Watson Tone Analyzer

```python
"""Example element
{
    "sentence": 'Mary did you know that your baby boy will one day walk on water?'
    "response": {'document_tone': {'tones': [{'score': 0.52647,
    'tone_id': 'joy',
    'tone_name': 'Joy'},
   {'score': 0.560098, 'tone_id': 'analytical', 'tone_name': 'Analytical'}]}}
}
"""
list_of_dict = [] #list of sentences with json of tone score in next to text
for i in range(9901, N):
    params = (
        ('version', '2017-09-21'),
        ('text', data[i])
    )
    response = requests.get('https://api.us-south.tone-analyzer.watson.cloud.ibm.com/instances/2a65b180-7db1-424e-a4ca-4339eba043be/v3/tone', params=params, auth=('apikey', 'LpOcRoOhRLlLxARDpgKpc-bnhQ3NqbZ1WFUeRB1o512S'))
    obj = {
        "sentence": data[i],
        "response": response.json()
    }
    list_of_dict.append(obj)
```

```python
len(list_of_dict)
```

```python
list_of_dict_cleaned = [elem for elem in list_of_dict if elem['response'] != {'document_tone': {'tones': []}}]
list_of_dict_cleaned = [elem for elem in list_of_dict_cleaned if elem['response'] != {'code': 403, 'error': 'Forbidden'}]
list_of_dict_cleaned = [elem for elem in list_of_dict_cleaned if elem['response'] != {'code': 400, 'sub_code': 'C00006', 'error': 'No text given'}]
list_of_dict_cleaned = [elem for elem in list_of_dict_cleaned if elem['response'] != ""]
```

```python
def splitbyemotion(emotion):
    emotion_list = []
    for elem in list_of_dict_cleaned:
        if elem['response'] != None and elem['response']['document_tone'] != None:
            for i in elem['response']['document_tone']['tones']:
                if i['tone_id'] == emotion:
                    emotion_list.append(elem)
    return emotion_list
```

```python
joy = splitbyemotion('joy')
anger = splitbyemotion('anger')
fear = splitbyemotion('fear')
sadness = splitbyemotion('sadness')
def sort_by_score(emotion_list, score):
    max_score = score
    new_list = []
    for elem in emotion_list:
        if elem['response']['document_tone']['tones'][0]['score'] > max_score:
            new_list.append({
                "sentence": elem['sentence'],
                "score": elem['response']['document_tone']['tones'][0]['score']
            })
    return new_list
```

```python
sort_by_score(anger, 0.7)
```

```python
def to_string(score_list):
    string = ""
    for elem in score_list:
        string += " " + elem['sentence']
    return string

def list_all_words(score_list):
    string = to_string(score_list)
    set_of_words = set(string.split(' '))
    return set_of_words

def word_count(score_list):
    list_of_words = list_all_words(score_list)
    list_with_count = []
    string = to_string(score_list)
    all_words = string.split(' ')
    for word in list_of_words:
        count = 0
        for i in all_words:
            if word == i:
                count+= 1
        list_with_count.append({
            "word": word,
            "count": count
        })
    return list_with_count
```

```python
list_joy = word_count(sort_by_score(joy, 0.5))
list_anger = word_count(sort_by_score(anger, 0.5))
list_fear = word_count(sort_by_score(fear, 0.5))
list_sadness = word_count(sort_by_score(sadness, 0.5))
```

```python
sorted(list_anger, key = lambda i: i['count'], reverse=True) 
```
